```python
import pickle
with open('svm_car_type_model.pkl', 'rb') as f:
    model = pickle.load(f)
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
```

```python
new_car = [[2.5, 5]]  # Example: EngineSize=2.5, Seats=5
new_car_scaled = scaler.transform(new_car)
predicted_type = model.predict(new_car_scaled)
print("Predicted Car Type for the new car:", predicted_type)
```
